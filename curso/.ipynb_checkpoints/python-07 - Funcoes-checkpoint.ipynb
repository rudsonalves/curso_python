## 7.9. Gerar uma Saída de Erro (sys.exit)

Explorando um pouco mais o pacote sys, empregar o sys.exit para sair do programa.


```

#!/bin/python3
#
from math import pi
import sys


def circulo(raio):
    return pi*raio**2


def quadrado(aresta):
    return aresta**2


def help():
    print('É necessário informar um raio/aresta.')
    print('\nSintaxe: {} <raio/aresta>\n'.format(sys.argv[0]))


if __name__ == '__main__':
    arg = sys.argv

    if len(arg) < 2:
        help()
        exit(1)
        
    if not arg[1].isnumeric():
        print('O argumento deve ser um valor numérico valido!')
        exit(2)
    
    
    raio = float(arg[1])

    area = circulo(raio)
    print(f'Área do círculo de raio {raio}m é {area:.2f}m²')

    area = quadrado(raio)
    print(f'Área do quadrado de aresta {raio}m é {area:.2f}m²')

```

É possível usar os códigos de erros padrões do sistema com o pacote `errno`:


```

#!/bin/python3
#
from math import pi
import sys
import errno

def circulo(raio):
    return pi*raio**2


def quadrado(aresta):
    return aresta**2


def help():
    print('É necessário informar um raio/aresta.')
    print('\nSintaxe: {} <raio/aresta>\n'.format(sys.argv[0]))


if __name__ == '__main__':
    arg = sys.argv

    if len(arg) < 2:
        help()
        exit(errno.EPERM)
        
    if not arg[1].isnumeric():
        print('O argumento deve ser um valor numérico valido!')
        exit(errno.EINVAL)
    
    raio = float(arg[1])

    area = circulo(raio)
    print(f'Área do círculo de raio {raio}m é {area:.2f}m²')

    area = quadrado(raio)
    print(f'Área do quadrado de aresta {raio}m é {area:.2f}m²')

```

### 7.9.1. Adicionar Cor na Mensagem de Erro


```

#!/bin/python3
#
from math import pi
import sys
import errno

ERRO = '\033[91m'
NORMAL = '\033[0m' 

def circulo(raio):
    return pi*raio**2


def quadrado(aresta):
    return aresta**2


def help():
    print('É necessário informar um raio/aresta.')
    print('\nSintaxe: {} <raio/aresta>\n'.format(sys.argv[0]))


if __name__ == '__main__':
    arg = sys.argv

    if len(arg) < 2:
        help()
        exit(errno.EPERM)
        
    if not arg[1].isnumeric():
        print(ERRO, 'O argumento deve ser um valor numérico valido!', NORMAL)
        exit(errno.EINVAL)
    
    raio = float(arg[1])

    area = circulo(raio)
    print(f'Área do círculo de raio {raio}m é {area:.2f}m²')

    area = quadrado(raio)
    print(f'Área do quadrado de aresta {raio}m é {area:.2f}m²')

```

### 7.9.2. Por as Cores em uma Classe


```

#!/bin/python3
#
from math import pi
import sys
import errno

class Color:
    ERRO = '\033[91m'
    NORMAL = '\033[0m' 

def circulo(raio):
    return pi*raio**2


def quadrado(aresta):
    return aresta**2


def help():
    print('É necessário informar um raio/aresta.')
    print('\nSintaxe: {} <raio/aresta>\n'.format(sys.argv[0]))


if __name__ == '__main__':
    arg = sys.argv

    if len(arg) < 2:
        help()
        exit(errno.EPERM)
        
    if not arg[1].isnumeric():
        print(Color.ERRO, 'O argumento deve ser um valor numérico valido!', Color.NORMAL)
        exit(errno.EINVAL)
    
    raio = float(arg[1])

    area = circulo(raio)
    print(f'Área do círculo de raio {raio}m é {area:.2f}m²')

    area = quadrado(raio)
    print(f'Área do quadrado de aresta {raio}m é {area:.2f}m²')

```


A vantagem deste modelo é que reduz a quantidade de objetos adicionados ao escopo global do programa, o que dificulta a superposição indesejada de objetos.

Adicione um `print(dir())` para mostrar isto.

Observe o erro `line too long (91 > 79 characters)Flake8(E501)`. Isto não é um erro no programa, mas sim no PEP8, um pacote de boas práticas no código instalado pelo Visual Studio Code.

Para corrigir quebre a linha e alinha com o início da linha superior.


```

#!/bin/python3
#
from math import pi
import sys
import errno

class Colors:
    ERRO = '\033[91m'
    NORMAL = '\033[0m' 

def circulo(raio):
    return pi*raio**2


def quadrado(aresta):
    return aresta**2


def help():
    print('É necessário informar um raio/aresta.')
    print('\nSintaxe: {} <raio/aresta>\n'.format(sys.argv[0]))


if __name__ == '__main__':
    arg = sys.argv

    if len(arg) < 2:
        help()
        exit(errno.EPERM)
        
    if not arg[1].isnumeric():
        print(Colors.ERRO, 
              'O argumento deve ser um valor numérico valido!', 
              Colors.NORMAL)
        exit(errno.EINVAL)
    
    raio = float(arg[1])

    area = circulo(raio)
    print(f'Área do círculo de raio {raio}m é {area:.2f}m²')

    area = quadrado(raio)
    print(f'Área do quadrado de aresta {raio}m é {area:.2f}m²')

```
